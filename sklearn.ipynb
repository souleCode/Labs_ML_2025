```python
from sklearn.linear_model import LinearRegression,SGDRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.datasets import make_regression
from matplotlib import pyplot as plt
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.datasets import make_regression
```

```python

x, y = make_regression(n_samples=100, n_features=1, noise=10)

# Afficher les premières lignes du dataset
plt.scatter(x, y)
```

```python
# Initialiser le modèle de régression linéaire
model = LinearRegression()

# Entraîner le modèle sur les données
model.fit(x, y)

# Faire des prédictions
y_pred = model.predict(x)
# Évaluer le modèle
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)
mae = np.mean(np.abs(y - y_pred))

print("MSE:", mse)
print("R² Score:", r2)
print("MAE:", mae)
```

```python
# Afficher y_pred et x
plt.scatter(x, y)
plt.plot(x, y_pred, color='red')
```

## Regression non lineaire

```python
# Transformer les données pour inclure des termes polynomiaux
poly = PolynomialFeatures(degree=2)
x_poly = poly.fit_transform(x)

# Initialiser et entraîner le modèle de régression linéaire sur les données transformées
model_poly = LinearRegression()
model_poly.fit(x_poly, y)
```

```python
ploy=PolynomialFeatures(degree=2)
x_poly=ploy.fit_transform(x)
model_poly = LinearRegression()
model_poly.fit(x_poly, y)
```

```python
# Faire des prédictions
y_poly_pred = model_poly.predict(x_poly)

# Évaluer le modèle
mse_poly = mean_squared_error(y, y_poly_pred)
r2_poly = r2_score(y, y_poly_pred)
mae_poly = np.mean(np.abs(y - y_poly_pred))

print("MSE (Polynomial):", mse_poly)
print("R² Score (Polynomial):", r2_poly)
print("MAE (Polynomial):", mae_poly)

# Afficher les résultats
plt.scatter(x, y, color='blue', label='Données réelles')
plt.plot(x, y_poly_pred, color='red', label='Prédictions polynomiales')
plt.legend()
plt.show()
```

```python
np.random.seed(0) # permet de reproduire l'aléatoire

X_multi, y_multi = make_regression(n_samples=100, n_features=2, noise = 10) # creation d'un dataset (x, y) linéaire

# Afficher x1 en abscisse et y en ordonnée
plt.scatter(X_multi[:, 0], y_multi, c='red', label='x1')

# Afficher x2 en abscisse et y en ordonnée
plt.scatter(X_multi[:, 1], y_multi, c='blue', label='x2')

# Ajouter des labels et une légende
plt.xlabel('x1 et x2')
plt.ylabel('y')
plt.legend()
```

```python
# Initialiser le modèle de régression linéaire pour les données multi-dimensionnelles
model_multi = LinearRegression()

# Entraîner le modèle sur les données multi-dimensionnelles
model_multi.fit(X_multi, y_multi)

# Faire des prédictions
y_multi_pred = model_multi.predict(X_multi)

# Évaluer le modèle
mse_multi = mean_squared_error(y_multi, y_multi_pred)
r2_multi = r2_score(y_multi, y_multi_pred)
mae_multi = np.mean(np.abs(y_multi - y_multi_pred))

print("MSE (Multi-dimensionnel):", mse_multi)
print("R² Score (Multi-dimensionnel):", r2_multi)
print("MAE (Multi-dimensionnel):", mae_multi)

# Afficher les résultats
plt.scatter(X_multi[:, 0], y_multi, color='blue', label='Données réelles x1')
plt.scatter(X_multi[:, 1], y_multi, color='green', label='Données réelles x2')
plt.scatter(X_multi[:, 0], y_multi_pred, color='red', label='Prédictions x1')
plt.scatter(X_multi[:, 1], y_multi_pred, color='orange', label='Prédictions x2')
plt.legend()
plt.show()
```

```python
# Changement des hyperparams
model_sgd= SGDRegressor(max_iter=1000, tol=1e-3)
model_sgd.fit(X_multi, y_multi)
y_multi_pred_sgd = model_sgd.predict(X_multi)
mse_multi_sgd = mean_squared_error(y_multi, y_multi_pred_sgd)
r2_multi_sgd = r2_score(y_multi, y_multi_pred_sgd)
mae_multi_sgd = np.mean(np.abs(y_multi - y_multi_pred_sgd))
print("MSE (Multi-dimensionnel):", mse_multi)
print("R² Score (Multi-dimensionnel):", r2_multi)
print("MAE (Multi-dimensionnel):", mae_multi)

# Afficher les résultats
plt.scatter(X_multi[:, 0], y_multi, color='blue', label='Données réelles x1')
plt.scatter(X_multi[:, 1], y_multi, color='green', label='Données réelles x2')
plt.scatter(X_multi[:, 0], y_multi_pred, color='red', label='Prédictions x1')
plt.scatter(X_multi[:, 1], y_multi_pred, color='orange', label='Prédictions x2')
plt.legend()
plt.show()
```
